# Heart Attact Prediction

**Membaca File Dataset**

```python
# Membaca File Dataset
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Portofolio/HeartAttackPrediction/Heart Predict')
df
```

**Mengecek Data Kosong**

```python
# Mengecek Datakosong
datakosong = df.isnull().sum()
datakosong[datakosong > 0]
```

```python
# Menghapus Datakosong
df = df.dropna() # Data diperbarui dengan membuang data yang kosong
```

**Visualisasi Data**

```python
# Visualisasi Data
import matplotlib.pyplot as plt
fig,ax = plt.subplots()

for output, d in df.groupby('output'):
  ax.scatter(d['trtbps'], d['chol'], label=output)

plt.legend(loc='upper left')
plt.title('Data Heart Attact')
plt.xlabel('Treat Blood Pleasure')
plt.ylabel('Cholesterol')
plt.grid(True)
plt.show()
```

**Preprocesing**

```python
# Preprocesing Dataset
import numpy as np

x_train = np.array(df[['chol', 'trtbps']])
y_train = np.array(df['output'])

print(f'X_Train:{x_train}')
print(f'Y_Train:{y_train}')
```

```python
from sklearn.preprocessing import LabelBinarizer # import sklearn labelbinarizer untuk melakukan processing

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train) # Mengubah Menjadi binary
print(f'Y_Train:\n{y_train}')
```

```python
y_train = y_train.flatten()
print(f'Y_Train:\n{y_train}')
```

**Proses Training Model Dengan KNN Clasification Model**

```python
from sklearn.neighbors import KNeighborsClassifier

k = 3
model = KNeighborsClassifier(n_neighbors=k)
model.fit(x_train, y_train)
```

**Prediksi Heart Attack**

```python
trtbps = 150
cholesterol = 200
x_new = np.array([trtbps, cholesterol]).reshape(1,-1)
x_new
```

```python
y_new = model.predict(x_new)
y_new
```

```python
lb.inverse_transform(y_new)
```

# Visualisasi Neresest Neighbours

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for output, d in df.groupby('output'):
  ax.scatter(d['trtbps'], d['chol'], label=output)

plt.scatter(trtbps,
            cholesterol,
            marker='s',
            color = 'red',
            label = 'Prediksi'
            )

plt.legend(loc='upper left')
plt.title('Penyebaran Data Trtbps dan Cholesterol')
plt.xlabel('Treat Blood Pleasure')
plt.ylabel('Cholesterol')
plt.grid(True)
plt.show()
```

# Evaluasi KNN Clasification Model

```python
x_test = np.array([[150,200],[100,70],[170,250],[200,300]])
y_test = lb.transform(np.array([1,0,0,1]))
print(f'X_Test:\n{x_test}')
print(f'Y_Test:\n{y_test}')
```

```python
y_pred = model.predict(x_test)
y_pred
```

# Acuraccy

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print(f'Accuracy: {acc}')
```

# Precission

```python
from sklearn.metrics import precision_score
prec = precision_score(y_test, y_pred)
print(f'Precision: {prec}')
```

# Recall

```python
from sklearn.metrics import recall_score
rec = recall_score(y_test, y_pred)
print(f'Recall: {rec}')
```

# F1 Score

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print(f'F1 Score: {f1}')
```

# Classification Report

```python
from sklearn.metrics import classification_report
cls = classification_report(y_test, y_pred)
print(f'Classification Report:\n{cls}')
```

# Mathews Correlation Coefficient (MCC)

```python
from sklearn.metrics import matthews_corrcoef
mcc = matthews_corrcoef(y_test, y_pred)
print(f'Matthews Correlation Coefficient: {mcc}')
```
